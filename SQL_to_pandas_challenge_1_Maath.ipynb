## 7. Select the job id and job description of all jobs, but display the columns with the names "id" and "description"

````python
(
jobs
    # rename the columns
    .rename(
        columns={
            'job_desc':'description', # rename job_desc to description
            'job_id':'id' # rename job_id to id
        }
    )
)
````

### HINT

```sql

	job_id AS id,
	job_desc AS description
FROM
	jobs;
```

## 8. Select all unique last names of authors (the same last name cannot appear more than once)

````python
(
authors
    .drop_duplicates(subset=['au_lname']) # drop duplicates
    [['au_lname']] # select only au_lname
)
````
